```python
import yfinance as yf
import pandas as pd
import os
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score
import matplotlib.pyplot as plt

# Check if CSV exists
if os.path.exists("nifty.csv"):
    nifty = pd.read_csv("nifty.csv", index_col=0)
else:
    nifty = yf.Ticker("^NSEI").history(period="max")
    nifty.to_csv("nifty.csv")

nifty
```

```python
# Data preparation
nifty.index = pd.to_datetime(nifty.index)
nifty.plot.line(y="Close", use_index=True)
del nifty["Dividends"]
del nifty["Stock Splits"]
```

```python
# Create target variable
nifty["Tomorrow"] = nifty["Close"].shift(-1)
nifty["Target"] = (nifty["Tomorrow"] > nifty["Close"]).astype(int)
nifty = nifty.loc["2007-09-17":].copy()
```

```python
# Model training
model = RandomForestClassifier(n_estimators=100, min_samples_split=250, random_state=1)
train = nifty.iloc[:-200]
test = nifty.iloc[-200:]
predictors = ["Close", "Volume", "Open", "High", "Low"]
```

```python
# Fit the model
model.fit(train[predictors], train["Target"])
```

```python

# Predictions and evaluation
preds = model.predict(test[predictors])
preds = pd.Series(preds, index=test.index)
print(precision_score(test["Target"], preds))
```

```python
# Combine results for visualization
combined = pd.concat([test["Target"], preds], axis=1)
combined.plot()
plt.figure(figsize=(12,6))

```

```python
# Function for prediction
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
# Backtesting function
def backtest(data, model, predictors, start=2000, step=250):
    all_predictions = []
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i + step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
# Perform backtesting
predictions = backtest(nifty, model, predictors)
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# Feature engineering with rolling averages
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = nifty.rolling(horizon).mean()

    # Create Close Ratio feature
    ratio_column = f"Close_Ratio_{horizon}"
    nifty[ratio_column] = nifty["Close"] / rolling_averages["Close"]

    # Create Trend feature
    trend_column = f"Trend_{horizon}"
    nifty[trend_column] = nifty.shift(1).rolling(horizon).sum()["Target"]

    new_predictors += [ratio_column, trend_column]
```

```python
# Drop rows with NaN values in relevant columns
nifty.dropna(subset=nifty.columns[nifty.columns != "Tomorrow"], inplace=True)
# Final model training with new predictors
model_final = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
# Perform final backtesting with new predictors
predictions_final = backtest(nifty, model_final, new_predictors)
precision_final = precision_score(predictions_final["Target"], predictions_final["Predictions"])

```

```python
# Display final predictions and precision score
print(predictions_final["Predictions"].value_counts())
print(f"Final Precision Score: {precision_final}")
```

```python
plt.figure(figsize=(12, 6))  # Adjust figure size as needed

# Plot Target values
plt.plot(predictions_final.index, predictions_final['Target'], label='Target', marker='o', linestyle='-')

# Plot Predictions values
plt.plot(predictions_final.index, predictions_final['Predictions'], label='Predictions', marker='s', linestyle='--')


# Customize the plot
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Comparison of Target and Predictions')
plt.legend()
plt.grid(True)  # Add a grid for better readability
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility

plt.tight_layout()  # Adjust layout to prevent overlapping elements
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(predictions_final["Target"], predictions_final["Predictions"]))
```
